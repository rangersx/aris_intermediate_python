#####2. Isi missing value pada kolom country dan children dengan data yang paling banyak muncul.

```python
columns = ["country","children"]
data[columns].fillna(data.mode().iloc[0])
```

####Buat kategori berdasarkan lead time: Short, Medium, dan Long

```python
import numpy as np
```

```python
data["CATEGORY"] = np.where(data['lead_time'] > 600, 'long', (np.where(data['lead_time'] < 300, 'short', 'medium')))
data
```

```python
kondisi = [
    (data['lead_time'] > 600),
    (data['lead_time'] > 300),
    (data['lead_time'] < 300)
]
     
values = ['Long','Medium', 'Short']
     
data["CATEGORY"] = np.select(kondisi, values)
data
```
