#####2. Isi missing value pada kolom country dan children dengan data yang paling banyak muncul.

```python
data['country'] = data['country'].fillna(data['country'].mode()[0])
```

```python
columns = ["country","children"]
tempContryChil = data[columns].fillna(data.mode().iloc[0])
```

```python
counts = tempContryChil['country'].value_counts().head()
country = tempContryChil['country'].value_counts().head().index.tolist()

tempframe = pd.DataFrame({'Country':country, 'Counts':counts})
tempframe.sort_values('Counts', ascending=False)

plt.bar('Country', 'Counts', data=tempframe)
plt.title('country graph')
plt.xlabel('country')
plt.ylabel('count')
plt.tight_layout()
plt.show()
```

```python
dataCountryChildren = tempContryChil['children'].value_counts()
labelCountryChildren = tempContryChil['children'].value_counts().index.tolist()
myexplode = [0,0,0,0.5,1.5]

plt.pie(dataCountryChildren, labels=labelCountryChildren, explode=myexplode, autopct = '%1.1f%%')
plt.title("children graph")
plt.tight_layout()
plt.show()
```

#####3. Isi missing value pada kolom agent dengan mean.

```python
dataAgent = data['agent'].value_counts().head()
labelAgent = data['agent'].value_counts().head().index.tolist()
myexplode = [0,0,0,0,0]

plt.pie(dataAgent, labels=labelAgent, explode=myexplode, autopct = '%1.1f%%')
plt.title("children graph")
plt.tight_layout()
plt.show()
```

####Buat kategori berdasarkan lead time: Short, Medium, dan Long

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# solution 1
data["CATEGORY"] = np.where(data['lead_time'] > 600, 'long', (np.where(data['lead_time'] < 300, 'short', 'medium')))
```

```python
# solution 1
tempLeadTime = []
tempLabelTime = ['Long', 'Medium', 'Short']
llong = len(data.loc[data["CATEGORY"] == "long"])
tempLeadTime.append(llong)
lmedium = len(data.loc[data["CATEGORY"] == "medium"])
tempLeadTime.append(lmedium)
lshort = len(data.loc[data["CATEGORY"] == "short"])
tempLeadTime.append(lshort)

tempLeadTime
```

```python
# solution 2
kondisi = [
    (data['lead_time'] > 600),
    (data['lead_time'] > 300),
    (data['lead_time'] < 300)
]
     
values = ['Long','Medium', 'Short']
     
data["CATEGORY"] = np.select(kondisi, values)
```

```python
# solution 2
tempLeadTime = data["CATEGORY"].value_counts()
tempLabelTime = data["CATEGORY"].value_counts().index.tolist()
tempLeadTime
```

```python
plt.pie(tempLeadTime, labels=tempLabelTime, autopct = '%1.1f%%')

plt.title("Lead Time")
plt.tight_layout()
plt.show()
```

####Filter semua data yang memiliki reservation_status_date di > 2017

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# reservation_status_date dari object ganti ke datetime
data["reservation_status_date"] = pd.to_datetime(data["reservation_status_date"])
data.dtypes
```

```python
data.dtypes
```

```python
#solution 1
filterLebih = data[data['reservation_status_date'].dt.year >= 2017]
```

```python
filterKurang = data[data['reservation_status_date'].dt.year < 2017]
```

```python
tempFilter = []
tempLabel = ['>= 2017', '< 2017']
countLebih = len(filterLebih)
tempFilter.append(countLebih)
countKurang = len(filterKurang)
tempFilter.append(countKurang)
```

```python
#solution 2
kondisi = [
    (data['reservation_status_date'].dt.year >= 2017),
    (data['reservation_status_date'].dt.year < 2017),
]
     
values = ['>= 2017','< 2017']
     
data["filter_year"] = np.select(kondisi, values)
```

```python
tempFilter2 = data["filter_year"].value_counts()
tempLabel = data["filter_year"].value_counts().index.tolist()
```
